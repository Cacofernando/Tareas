```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from datetime import date
from matplotlib.pyplot import figure
```

Investigando, se obtiene un dataset proveniente del Departamento de Estadísticas e Información de Salud (DEIS), el cual, contiene defunciones por causa de muerte entre los años 2016 - 2019, con frecuencia de actualización semanal (https://deis.minsal.cl/#datosabiertos)

Se abre el abre el archivo .csv mediante pandas.

```python
Defunciones = pd.read_csv('DEFUNCIONES_FUENTE_DEIS_2016_2021_22042021.csv', sep = ';', encoding = 'latin-1')
```

Sin embargo, como el dataset viene sin encabezado, no obstante, mediante el diccionario del archivo .csv, se agregan los nombres de las columnas y así, agregando el encabezado al dataset 

```python
Defunciones.columns = ['ANO_DEF','FECHA_DEF','GLOSA_SEXO','EDAD_TIPO','EDAD_CANT','CODIGO_COMUNA_RESIDENCIA','GLOSA_COMUNA_RESIDENCIA','GLOSA_REG_RES','DIAG1','CAPITULO_DIAG1','GLOSA_CAPITULO_DIAG1','CODIGO_GRUPO_DIAG1','GLOSA_GRUPO_DIAG1','CODIGO_CATEGORIA_DIAG1','GLOSA_CATEGORIA_DIAG1','CODIGO_SUBCATEGORIA_DIAG1','GLOSA_SUBCATEGORIA_DIAG1','DIAG2','CAPITULO_DIAG2','GLOSA_CAPITULO_DIAG2','CODIGO_GRUPO_DIAG2','GLOSA_GRUPO_DIAG2','CODIGO_CATEGORIA_DIAG2','GLOSA_CATEGORIA_DIAG2','CODIGO_SUBCATEGORIA_DIAG2','GLOSA_SUBCATEGORIA_DIAG2']
```

Por otro lado, como se pide el tratamiento de "muertes" por semana, se agregará una nueva columna llamada "SEMANA" para realizar el ánalisis descriptivo

```python
Defunciones.insert(2, 'SEMANA', 0)
```

El Dataset quedará de esta forma, incluyendo las defunciones de todos los casos de muerte


```python
Defunciones
```

Como la Fecha está en fomato String, es necesario transformar el contenido a fecha y por otro lado, a través de los datos de fecha, se extraerá el n° de semana con respecto a la fecha respectiva

```python
Defunciones['FECHA_DEF'] = pd.to_datetime(Defunciones['FECHA_DEF'])
```

```python
Defunciones['SEMANA'] = Defunciones['FECHA_DEF'].dt.week
```

```python
Defunciones
```

Se obtiene las defunciones de casos totales desde 2020 a 2021

```python
Defunciones_Totales = Defunciones[Defunciones.ANO_DEF.isin([2020,2021])]
```

```python
Defunciones_Totales
```

A través del nuevo dataset de casos totales, se filtrará aquellos casos de Gripe(influenza)

```python
Defunciones_Totales_Gripe = Defunciones_Totales[Defunciones_Totales.GLOSA_GRUPO_DIAG1.isin(["Influenza [gripe] y neumonía"])]
```

```python
Defunciones_Totales_Gripe
```

```python
Defunciones_Totales_COVID = Defunciones_Totales[Defunciones_Totales.GLOSA_SUBCATEGORIA_DIAG1.isin(['COVID-19 virus identificado', 'COVID-19 virus no identificado'])]
```

```python
Defunciones_Totales_COVID
```

Después, se generarán listas en blanco para cada tipo de casos de defunción para armar el histograma

```python
Frecuencia_Semanal_Casos_Totales = []
Frecuencia_Semanal_Casos_Gripe = []
Frecuencia_Semanal_Casos_COVID = []
```

```python
for i in Defunciones_Totales['SEMANA']:
    Frecuencia_Semanal_Casos_Totales.append(i);
    
for j in Defunciones_Totales_Gripe['SEMANA']:
    Frecuencia_Semanal_Casos_Gripe.append(j);
    
for k in Defunciones_Totales_COVID['SEMANA']:
    Frecuencia_Semanal_Casos_COVID.append(k);
    
```

```python
Intervalos_Casos_Totales = range(min(Frecuencia_Semanal_Casos_Totales), max(Frecuencia_Semanal_Casos_Totales) )
```

```python
Intervalos_Casos_Gripe = range(min(Frecuencia_Semanal_Casos_Gripe), max(Frecuencia_Semanal_Casos_Gripe))
```

```python
plt.hist(x = Frecuencia_Semanal_Casos_Totales, bins = Intervalos_Casos_Totales, color = 'b', rwidth = 0.7)
plt.title('Histograma de Defunciones de Casos Totales (Desde el año 2019 al 2020)')
plt.xlabel('Semanas')
plt.ylabel('Frencuencia de Defunción')
plt.xticks(Intervalos_Casos_Totales)
plt.show()
```

```python
plt.hist(x = Frecuencia_Semanal_Casos_Gripe, bins = Intervalos_Casos_Gripe, color = '#F2AB6D', rwidth = 0.85)
plt.title('Histograma de Defunciones por Casos de Gripe por Semana (Desde el año 2019 al 2020)')
plt.xlabel('Semanas')
plt.ylabel('Frencuencia de Defunción por Gripe')
plt.xticks(Intervalos_Casos_Gripe)
plt.show()
```

```python
Intervalos_Casos_COVID = range(min(Frecuencia_Semanal_Casos_COVID), max(Frecuencia_Semanal_Casos_COVID))
```

```python
plt.hist(x = Frecuencia_Semanal_Casos_COVID, bins = Intervalos_Casos_COVID, color = 'tab:red', rwidth = 0.85)
plt.title('Histograma de Defunciones por Casos de COVID por Semana (Desde el año 2019 al 2020)')
plt.xlabel('Semanas')
plt.ylabel('Frencuencia de Defunción por COVID')
plt.xticks(Intervalos_Casos_COVID)
plt.show()
```

No obstante, si comparamos los casos totales vs COVID, y los casos de gripe vs COVID

```python
plt.hist(x = Frecuencia_Semanal_Casos_Totales, bins = Intervalos_Casos_Totales, color = 'b', rwidth = 0.7)
plt.hist(x = Frecuencia_Semanal_Casos_COVID, bins = Intervalos_Casos_COVID, color = 'tab:red', rwidth = 0.7)
plt.title('Histograma de Defunciones de Casos Totales vs Defunciones por Casos COVID por Semana (Desde el año 2019 al 2020)')
plt.xlabel('Semanas')
plt.ylabel('Frencuencia de Defunción')
plt.xticks(Intervalos_Casos_Totales)
plt.show()
```

```python
plt.hist(x = Frecuencia_Semanal_Casos_COVID, bins = Intervalos_Casos_COVID, color = 'tab:red', rwidth = 0.7)
plt.hist(x = Frecuencia_Semanal_Casos_Gripe, bins = Intervalos_Casos_Gripe, color = '#F2AB6D', rwidth = 0.7)
plt.title('Histograma de Defunciones por Casos de Gripe vs Casos de COVID por Semana (Desde el año 2019 al 2020)')
plt.xlabel('Semanas')
plt.ylabel('Frencuencia de Defunción por Gripe')
plt.xticks(Intervalos_Casos_Gripe)
plt.show()
```

## Revisando los histogramas

Si comparamos los tres histogramas, podemos observar en primer lugar, que datos de los fallecidos por gripe(Neumonía) y los fallecidos por COVID-19, se comportan de manera distinta, ya que en las semanas N° 16 y 19, hubo un deceso de caso debido a la primera cuarentena aplicada a nivel de país. También, cabe recalcar que el histograma de los casos totales se relaciona más con el histograma de fallecidos por COVID, ya que, desde el año 2020 hasta la actualidad, los casos de fallecidos son en mayor grado por Covid que por la Gripe(Neumonía), por lo tanto, se puede concluir que el comportamiento del COVID es distinto a la gripe común.
